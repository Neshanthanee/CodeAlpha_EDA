```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the cleaned dataset
file_path = "Data.csv" 
df = pd.read_csv(file_path)
```

```python

print("--- Data Structure ---")
print(df.head())           
print("\n")
print(df.info())          
print("\n")
print(df.describe())       
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Price (GBP)', y='Star Rating', data=df)
plt.title('Price vs. Star Rating of Books')
plt.xlabel('Price (GBP)')
plt.ylabel('Star Rating')
plt.grid(True)
plt.show()
```

```python

correlation = df['Price (GBP)'].corr(df['Star Rating'])
print(f"\nCorrelation between Price and Rating: {correlation:.2f}")

plt.figure(figsize=(7, 5))
sns.countplot(x='Star Rating', data=df, order=sorted(df['Star Rating'].unique()))
plt.title('Distribution of Star Ratings')
plt.xlabel('Star Rating')
plt.ylabel('Number of Books')
plt.show()
```

```python

plt.figure(figsize=(6, 8))
sns.boxplot(y=df['Price (GBP)'])
plt.title('Box Plot of Book Prices (Outliers)')
plt.ylabel('Price (GBP)')
plt.show()
```
